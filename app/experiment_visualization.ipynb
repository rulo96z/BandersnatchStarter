```python
import data
from altair import Chart
```

```python
db = data.Database()
df = db.dataframe()
df
```

```python
# Create the chart
Chart(df).mark_circle(size=60).encode(
    x='Level',
    y='Energy',
    color='Rarity',
    tooltip=['Level', 'Rarity', 'Damage', 'Health', 'Energy', 'Sanity']

).properties(
    title='Energy by Level for Rarity',
    height=400,
    width=350,
    padding=50,
    background='#252525'

).configure_title(
    fontSize=25,
    color='white',
    offset=25

).configure_axis(
    labelColor='white',
    titleColor='white',
    titlePadding=20,
    domainColor='white',  
    gridOpacity=0.1

).configure_legend(
    labelColor='white',
    titleColor='white'

).configure_view(
    stroke='transparent'

)#.interactive()
```

```python
# Create a sample chart
chart = Chart(df).mark_point().encode(
    x='Level',
    y='Energy',
    color='Rarity',
    tooltip=['Level', 'Rarity', 'Damage', 'Health', 'Energy', 'Sanity'])

# Convert the chart to JSON
chart_json = chart.to_json()

# Print the JSON representation
print(chart_json)
```
